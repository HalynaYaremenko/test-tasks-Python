1. –ø–µ—Ä–µ—Ç–≤–æ—Ä–∏ –ø–µ—Ä—à—É –ª—ñ—Ç–µ—Ä—É —É –≤–µ–ª–∏–∫—É `(capitalize())`

````python
sentence = "learning python is easy and fun"

print (sentence.capitalize())

res = 0
for i in sentence:
    if i == "a":
        res += 1
print(res)

a = sentence.split()
for j in a:
    if j == "fun":
        print("True")
````
